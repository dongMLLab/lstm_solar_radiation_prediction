```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
combined = pd.read_csv("combined.csv")

combined["date"] = pd.to_datetime(combined["date"], format="%Y-%m-%d %H:%M:%S")
combined.index = combined["date"]

combined["year"] = combined["date"].dt.year
combined["month"] = combined["date"].dt.month
combined["day"] = combined["date"].dt.day
combined["hour"] = combined["date"].dt.hour
combined
```

```python
# TODO 일단위, 월단위, 주단위 시각화 + 상관관계 + 회귀식
```

```python
plt.title("Solar Radiation (2017.1.1 ~ 2023.10.18)")

plt.plot(combined["solar_radiation"],)
plt.xlabel("Date")
plt.ylabel("Solar Radiation")
plt.savefig("./graph/whole_solar_radiation.jpeg")
plt.show()
```

```python
plt.title("Daily Solar Radiation Sample (2022.10.18)")

plt.plot(combined[(combined["day"] == 18) & (combined["year"] == 2022) & (combined["month"] == 10)]["hour"], 
         combined[(combined["day"] == 18) & (combined["year"] == 2022) & (combined["month"] == 10)]["solar_radiation"], 
         label="Solar Radiation",
         c="r")

plt.ylabel("Solar Radiation")
plt.ylim(top=3.0, bottom=0)
plt.legend(loc="best")
plt.xlabel("Hour")
plt.savefig("./graph/daily_solar_radiation.jpeg")
plt.show()
```
