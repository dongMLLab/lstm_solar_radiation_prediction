```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("combined.csv")

data["date"] = pd.to_datetime(data["date"], format="%Y-%m-%d %H:%M:%S")
data = data.drop("date.1", axis=1)
data.index = data["date"]

data = data.fillna(0)

data
```

```python
hour_range = np.arange(9, 19)
hour_range
```

```python
# 활동 시간 중
selected_data = data[data["date"].dt.hour.isin(hour_range)]
selected_data = selected_data.drop("date", axis = 1)
selected_data
```

```python
# 활동 시간에 비가 온 날들
selected_data2 = data[(data["rain"] > 0) & (data["date"].dt.hour.isin(hour_range))]
selected_data2 = selected_data2.drop("date", axis = 1)
selected_data2
```

```python
# 전체 데이터 그대로
pearson_selected_corr0 = data.corr("pearson")

pearson_selected_corr0
```

```python
# 전체 데이터에서 비가 온 데이터만 -> 필요없을 듯.
pearson_selected_corr0_1 = data[data["rain"]>0].corr("pearson")
# spearman_selected_corr0_1 = data[data["rain"]>0].corr("spearman")
pearson_selected_corr0_1
```

```python
# 9시 ~ 18시
pearson_selected_corr = selected_data.corr("pearson")

pearson_selected_corr
```

```python
# 9~18시 중 비가 온 날만
pearson_selected_corr2 = selected_data2.corr("pearson")
# spearman_selected_corr2 = selected_data2.corr("spearman")

pearson_selected_corr2
```

```python
# 전체 데이터에 대해
pearson0 = pearson_selected_corr0[(abs(pearson_selected_corr0["solar_radiation"]) >= 0.1) & 
                                  (abs(pearson_selected_corr0["solar_radiation"] < 1))]["solar_radiation"].sort_values(ascending=False)

frame0 = pd.DataFrame()
frame0["Pearson"] = pearson0

frame0
```

```python
# 전체 데이터에 대해
pearson0 = pearson_selected_corr0[(abs(pearson_selected_corr0["solar_radiation"]) >= 0.2) & 
                                  (abs(pearson_selected_corr0["solar_radiation"] < 1))]["solar_radiation"].sort_values(ascending=False)

frame0 = pd.DataFrame()
frame0["Pearson"] = pearson0

frame0
```

```python
# 전체 데이터에서 비가 온 데이터만
pearson0_1 = pearson_selected_corr0_1[(abs(pearson_selected_corr0_1["solar_radiation"]) >= 0.1) & 
                                      (abs(pearson_selected_corr0_1["solar_radiation"] < 1))]["solar_radiation"].sort_values(ascending=False)

frame0_1 = pd.DataFrame()

frame0_1["Pearson"] = pearson0_1

frame0_1
```

```python
# 활동시간 중
pearson = pearson_selected_corr[(abs(pearson_selected_corr["solar_radiation"]) >= 0.1) & 
                                (abs(pearson_selected_corr["solar_radiation"] < 1))]["solar_radiation"].sort_values(ascending=False)

frame = pd.DataFrame()

frame["Pearson"] = pearson

frame
```

```python
# 활동 시간에 비가 온 날들
pearson2 = pearson_selected_corr2[(abs(pearson_selected_corr2["solar_radiation"]) >= 0.1) &  (abs(pearson_selected_corr2["solar_radiation"] < 1))]["solar_radiation"].sort_values(ascending=False)
# spearman2 = spearman_selected_corr2[(abs(spearman_selected_corr2["solar_radiation"]) >= 0.2) &  (abs(spearman_selected_corr2["solar_radiation"] < 1))]["solar_radiation"].sort_values(ascending=False)

frame2 = pd.DataFrame()
frame2["Pearson"] = pearson2
# frame2["Spearman"] = spearman2

frame2
```
